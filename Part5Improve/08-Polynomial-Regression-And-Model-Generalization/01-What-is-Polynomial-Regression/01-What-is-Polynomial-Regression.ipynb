# 多项式回归

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
```

```python
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size=100)
```

```python
plt.scatter(x, y)
plt.show()
```

## 线性拟合

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X, y)
```

```python
y_predict = lin_reg.predict(X)
```

```python
plt.scatter(x, y)
plt.scatter(x, y_predict, color='r')
plt.show()
```

可以看到线性拟合的效果很差

## 解决方案：添加一个特征

```python
X2 = np.hstack([X, X**2])
```

```python
X2.shape
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X2, y)
y_predict2 = lin_reg2.predict(X2)
```

```python
plt.scatter(x, y)
plt.scatter(np.sort(x), y_predict2[np.argsort(x)], color='r')
plt.show()
```
